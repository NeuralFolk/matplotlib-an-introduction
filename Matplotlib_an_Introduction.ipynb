```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

image = plt.imread("/content/4870863_carP_l_1_jpg20211112-1405-ncsm71.jpg")
plt.imshow(image)
```

```python
x = np.arange(10, 20)
y = np.arange(100, 110)
```

```python
plt.plot(x, y, ls = 'dashed', color = '#000000', marker = 's', ms = 10, mec = 'red', mfc = 'green')
plt.title("Title")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(axis = 'y')
plt.show()
```

```python
x = np.random.normal(100, 15, (9, 9))
y = np.random.normal(50, 10, (9, 9))
```

```python
plt.scatter(x, y)
plt.show()
```

```python
x = np.arange(6)
labels = ["Apple", "Banana", 'Orange', 'Lichi', "Mango", 'Blackberry']
plt.pie(x, labels = labels)
plt.show()
```

```python
x = np.random.normal(100, 10, 100)
plt.hist(x)
plt.show()
```
